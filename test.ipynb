```python
import sys, os

script_dir = os.path.dirname(os.path.abspath('__file__'))
module_dir = os.path.join(script_dir, "build")

sys.path.append(module_dir)


import ODESolvers as ode
import math
```

```python
def func(y, t):
    return -y - math.exp(-t) * math.sin(t)
```

```python
y = 0
h = 0.1
t = 0
y_vals = []
t_vals = []
while t <= 20:
    y_vals.append(y)
    t_vals.append(t)
    result = ode.DormandPrince(0.00005).next_step(y, h, func, t)
    if result.h < 1e-6:
        break
    if result.accepted:
        y = result.y
        t += h
    h = result.h
 
```

```python
import matplotlib.pyplot as plt

plt.plot(t_vals, y_vals, label="Euler")
y_acc = []
for t in t_vals:
    y_acc.append(-t * math.exp(-t) * math.cos(t))
plt.plot(t_vals, y_acc, label="Exact")
plt.legend()
plt.show()
```
