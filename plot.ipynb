```python
import matplotlib.pyplot as plt
```

```python
t1 = []
y1 = []
t2 = []
y2 = []
with open("example.txt") as f:
   for line in f.readlines():
      _t, _y = line.split(",")
      t1.append(float(_t))
      y1.append(float(_y))
with open("example2.txt") as f:
   for line in f.readlines():
      _t, _y = line.split(",")
      t2.append(float(_t))
      y2.append(float(_y))

from scipy.integrate import solve_ivp
import math

# f = log(abs(y)) - cos(t)

def dydt(t, y):
    return math.log(abs(y)) - math.cos(t)

y0 = [1]
t_span = (0, 3)

solution = solve_ivp(dydt, t_span, y0)


plt.plot(solution.t, solution.y[0])

plt.plot(t1, y1)
plt.plot(t2, y2)
plt.show()
```
