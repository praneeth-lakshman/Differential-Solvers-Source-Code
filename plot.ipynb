```python
import matplotlib.pyplot as plt
```

```python
t1 = []
y1 = []
t2 = []
y2 = []
t3 = []
y3 = []
t4 = []
y4 = []
tm = []
ym = []
with open("example.txt") as f:
   for line in f.readlines():
      _t, _y = line.split(",")
      t1.append(float(_t))
      y1.append(float(_y))
with open("example1.txt") as f:
   for line in f.readlines():
      _t, _y = line.split(",")
      t2.append(float(_t))
      y2.append(float(_y))
with open("example2.txt") as f:
   for line in f.readlines():
      _t, _y = line.split(",")
      t3.append(float(_t))
      y3.append(float(_y))
with open("example3.txt") as f:
   for line in f.readlines():
      _t, _y = line.split(",")
      t4.append(float(_t))
      y4.append(float(_y))
with open("matlab.txt") as f:
   for line in f.readlines():
      _t, _y = line.split(",")
      tm.append(float(_t))
      ym.append(float(_y)) 

from scipy.integrate import solve_ivp
import math

# f = log(abs(y)) - cos(t)

def dydt(t, y):
    return ((y-1)**2) * ((t-1)**2)

y0 = [0]
t_span = (0, 5)

solution = solve_ivp(dydt, t_span, y0)


plt.plot(solution.t, solution.y[0], label="Scipy")

plt.plot(t1, y1, label="Euler")
plt.plot(t2, y2, label="RK4")
plt.plot(t3, y3, label="RKF45")
plt.plot(t4, y4, label="Dormand Prince")
plt.plot(tm, ym, label="MatLab")
plt.legend()
plt.show()
```

```python
from scipy.integrate import solve_ivp
import math

# f = log(abs(y)) - cos(t)

def dydt(t, y):
    return math.log(abs(y)) - math.cos(t)

y0 = [1]
t_span = (0, 3)

solution = solve_ivp(dydt, t_span, y0)


plt.plot(solution.t, solution.y[0])
```
